## Framework
1.  Define the problem
- Problems before requirements
- Requirements before solutions
- Solutions before design
- Design before technology
2.  Gather the Data
3.  Prepare Data for Consumption, aka Data Wrangling
- Implementing data structures for storage and processing
- Developing data governmance standards for quality control
- Data extraction (ETL - Extract, Transform, Load)
- Data cleaning to identify aberrant, missing, or outlier data points
4.  Perform Exploratory Analysis
- Descriptive and graphical statistics
- Look for potential problems, patterns, classifications, correlations, and comparisons
- Data categorization (qualitative, quantitative)
- Use this step to select the correct hypothesis set or data model.
5.  Model Data
- LIke descriptive and inferential statistics, data modeling can summarize or predict.
6.  Validate and Implement
- Determine whether our model overfit, generalized, or underfit the dataset.
7.  Optimize and Strategize
- Iterate back through the process to improve.

### Import Data
- Two data sets, the training and test data
- Twelve columns of data
- 'Survived' is the dependent variable
- 'PassengerId' and 'Ticket' are not useful at predicting the dependent variable
- 'SibSp' is number of siblings or spouse
- 'Parch' is number of parents or children
- 'Embarked' is the port of embarkation
- 'Cabin' may be useful, but only with additional information about how the cabin numbers correlate to locations on the ship.

```python
data_raw = pd.read_csv('../train.csv')
data_val = pd.read_csv('../test.csv')
print (data_raw.info())
data_raw.head()
data1 = data_raw.copy(deep=True)
data_cleaner = [data1, data_val]
```

### Look for Null Values
- 'Cabin' has lots of null values, but it's not a useful piece of information anyway.
- 'Age' has many null values; we may replace the null with the mean age, or mean for that Pclass or Sex.
- 'Embarked' has some null values in the Training set; we may replace those with the mode port.
- 'Fare' has one null value in the test set; we may replace it with the mean, median, or mode fare for that Pclass.  

```python
print ('Train columns with null values')
print (data1.isnull().sum())
print ('-*'*20)
print ('Test/Validation columns with null values')
print (data_val.isnull().sum())
print ('=*'*20)
data_raw.describe (include='all')
```

### Complete or delete missing values in train and/or test/validation dataset.

```python
for dataset in data_cleaner:
    # Complete missing age with median
    dataset['Age'].fillna(dataset['Age'].median(), inplace=True)
    # Complete embarked with mode
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace=True)
    # Complete missing fare with median
    dataset['Fare'].fillna(dataset['Fare'].median(), inplace=True)
# Delete the PassengerId, Cabin, and Ticket features/columns
## from the training dataset
data1.drop(['PassengerId','Cabin','Ticket'], axis=1, inplace=True)
for dataset in data_cleaner:
    print (dataset.isnull().sum())


```

## Create:  Feature engineering for training and validation datasets

```python
for dataset in data_cleaner:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
    dataset['IsAlone'] = 1
    dataset['IsAlone'].loc[dataset['FamilySize']>1] = 0
    dataset['Title'] = dataset['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
    dataset['FareBin'] = pd.qcut(dataset['Fare'], 4)
    dataset['AgeBin'] = pd.cut(dataset['Age'].astype(int), 5)
#    print (dataset.head())
# Clean up rare title names
stat_min = 10
title_names = (data1['Title'].value_counts() < stat_min)
data1['Title'] = data1['Title'].apply(lambda x: 'Misc' if title_names.loc[x] == True else x)
print (data1['Title'].value_counts())
print ('=*'*20)

data1.info()
data_val.info()
data1.sample(10)
```

### Convert Formats
- Convert categorical data to dummy variables for mathematical analysis.

```python
# Code categorical data
label = LabelEncoder()
for dataset in data_cleaner:
    for code in [
        ['Sex_Code', 'Sex'],
        ['Embarked_Code', 'Embarked'],
        ['Title_Code', 'Title'],
        ['AgeBin_Code', 'AgeBin'],
        ['FareBin_Code', 'FareBin']
    ]:
        dataset[code[0]] = label.fit_transform(dataset[code[1]])
print (data1.sample(10))
```

### Define our $x$ (independent variables) and $y$ (dependent/target/outcome/response) variables for data modeling.  
- `Dummy variables' here transform categories into one-hot encoding.  

```python
Target = ['Survived']
data1_x = ['Sex', 'Pclass', 'Embarked', 'Title', 'SibSp', 'Parch', 'Age', 'Fare', 'FamilySize', 'IsAlone']
data1_x_calc = ['Sex_Code', 'Pclass', 'Embarked_Code', 'Title_Code', 'SibSp', 'Parch', 'Age', 'Fare']
data1_xy = Target + data1_x
print ('Original X Y: ', data1_xy, '\n')

data1_x_bin = ['Sex_Code', 'Pclass', 'Embarked_Code', 'Title_Code', 'FamilySize', 'AgeBin_Code', 'FareBin_Code']
data1_xy_bin = Target + data1_x_bin
print ('Bin X Y: ', data1_xy_bin, '\n')

# Define x and y variables for dummy features original
data1_dummy = pd.get_dummies(data1[data1_x])
data1_x_dummy = data1_dummy.columns.tolist()
data1_xy_dummy = Target + data1_x_dummy
print ('Dummy X Y: ', data1_xy_dummy, '\n')

data1_dummy.sample(10)[['Sex_male']]
```

### Split Training and Testing Data

```python
train1_x, test1_x, train1_y, test1_y = model_selection.train_test_split(
data1[data1_x_calc], data1[Target], random_state = 0)

train1_x_bin, test1_x_bin, train1_y_bin, test1_y_bin = model_selection.train_test_split(
data1[data1_x_bin], data1[Target], random_state = 0)

train1_x_dummy, test1_x_dummy, train1_y_dummy, test1_y_dummy = model_selection.train_test_split(
data1_dummy[data1_x_dummy], data1[Target], random_state = 0)

print ('Data1 Shape: {}'.format(data1.shape))
print ('Train1 Shape: {}'.format(train1_x.shape))
print ('Test1 Shape: {}'.format(test1_x.shape))

train1_x_bin.head()
```
