```python
write = open('AAP2.bib', 'w')
for index, ser in A.iterrows():
    write.write('@%s{%s,\n' % (str(ser['ENTRYTYPE']), str(ser['ID'])))
    for item in ['title', 'journal', 'volume', 'number', 'pages', 'year', 'author', 'keywords', 'abstract', 'institution', 'annotation', 'addendum']:
        write.write('%s = {%s},\n' % (item, str(ser[item])))
    write.write('}\n\n')
    print (ser.author)
write.write('\n')


    
```
