```python
import bibtexparser
import pandas as pd
import numpy as np
```

```python
with open('Anystyle.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)
```

```python
P = pd.DataFrame(bib_database.entries)
P
```

```python
P['author'] = P['author'].fillna('None')
A = [ x.split(' and ') for x in P['author'].tolist() ]
B = [item for sublist in A for item in sublist]
C = {x:B.count(x) for x in B}
D = dict(sorted(C.items(), key=lambda item: item[1], reverse=True))
D
```

```python
A = P['author'].tolist()
C = {x:A.count(x) for x in A}
D = dict(sorted(C.items(), key=lambda item: item[1], reverse=True))
D
```

```python
A = P['title'].tolist()
C = {x:A.count(x) for x in A}
D = dict(sorted(C.items(), key=lambda item: item[1], reverse=True))
D
```

```python
A = P.groupby(['title', 'author','date']).size().reset_index(name='counts')
print (len(A))
A = A[A['counts']>5]
A.sort_values(by=['counts'], ascending=False, inplace=True)
A
```
